```python
import wikipedia as wp 
import pandas as pd
from tqdm import tqdm

import json
import re
```

```python
def save_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f)

def load_json(filename):
    with open(filename, 'r') as f:
        return json.load(f)
    
class wiki:
    @staticmethod
    def get_page(title):
        try:
            page = wp.page(title)
            print(f"Successfully downloaded article: {page.title}")
            return page
        except wp.PageError:
            print(f"Page '{title}' does not exist.")
            return None
    
    @staticmethod
    def get_content(title):
        page = wiki.get_page(title)
        return page.content if page is not None else ""
    
```

```python

article_title = "List of current members of the United States House of Representatives"
house_of_reps_page = wiki.get_page(article_title)
```

```python
try: 
    html = house_of_reps_page.html().encode("UTF-8")
    df = pd.read_html(html)[6]  
except IndexError:
    html = house_of_reps_page.html().encode("UTF-8")
    df = pd.read_html(html)[0]
df
```

```python
members_df = df
republicans_names=members_df[members_df['Party.1']=='Republican']['Member'].tolist()
republicans_names
```

```python
democrats_names=members_df[members_df['Party.1']=='Democratic']['Member'].tolist()
democrats_names
```

```python
from bs4 import BeautifulSoup
# tables = pd.read_html(html)

soup = BeautifulSoup(html, "html.parser")

# Find the table (adjust if necessary)
tables = soup.find_all("table", {"class": "wikitable"})
table = tables[5]

name_person_lookup = {}
for row in table.find_all("tr")[1:]:  # Skip header row
    first_column = row.find("b")  # Adjust if necessary
    if first_column and first_column.a:
        person_name = first_column.a.text.strip()
        person_link = first_column.a["href"]
        name_person_lookup.update({person_name : person_link})

# Print or save results
for name, link in name_person_lookup.items():
    print(f"{name}: {link}")
```

```python
for name , link in name_person_lookup.items():
    link = re.sub(r"^/wiki/", "", link).replace("_", " ")
    name_person_lookup.update({name : link})

name_person_lookup
```

```python
republicans_links = {name: name_person_lookup[name] for name in republicans_names if name in name_person_lookup}
republicans_links
```

```python
democrats_links = {name: name_person_lookup[name] for name in democrats_names if name in name_person_lookup}
```

```python
republicans_article_names = [republicans_links[name] for name in republicans_links]
democrats_article_names = [democrats_links[name] for name in democrats_links]
members_of_congress = {"Republicans": republicans_article_names, "Democrats": democrats_article_names}  
save_json(members_of_congress, "members_of_congress_wiki_pages.json")
```
